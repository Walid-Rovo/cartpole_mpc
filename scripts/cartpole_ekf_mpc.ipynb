## Pendulum Simulator Demo
The simulator used in this demo is heavily inspired by the Python package Gymnasium, specifically its included classic control environments.

```python
import numpy as np
from cartpole_sim import PendulumOnCart

my_pendulum = PendulumOnCart(render=True)

SIM_TIME = 20  # seconds
SIM_STEP = 0.01
N_SIM_STEPS = int(SIM_TIME/SIM_STEP)
FORCE_PERIOD = 1
FORCE_MAGNITUTUDE = 0.01  # newtons

# sin force
t_vec = np.arange(stop=SIM_TIME/4, step=SIM_STEP)
force_vec = FORCE_MAGNITUTUDE * np.sin(t_vec * (2 / FORCE_PERIOD * np.pi))

import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
plt.title("Force (N) vs Time (s)")
ax.plot(t_vec, force_vec, 'b-')
plt.show()
```

```python
my_pendulum.reset()

#
for f in force_vec:
    my_pendulum.step(action=f)

my_pendulum.close()
```
